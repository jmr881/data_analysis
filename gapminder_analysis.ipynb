Extract information

```python
def pop_year(df, year): 
    '''Function that calculates total pop per continent for a year and Input: a dataframe, year Output: resulting dataframe'''
    assert type(year) is int, "Argument is not a number. Try again"
    assert year > 1950, "Year too early"
    population_data = df[(df.year == year)] [["continent", "pop" ] ].groupby("continent").sum().sort("pop")
    return population_data
print pop_year(gap_data, 1952)
```

```python
def pop_plot(df, year):
    year_string = str(year)
    types = list(df.index)
    y_pos = np.arange(len(types))
    plt.barh(y_pos, df["pop"], align='center')
    plt.yticks(y_pos, types)
    plt.xlabel('continent')
    plt.ylabel('population')
    plt.title('World populations in '+year_string)
    plt.show()
pop_plot(pop_year(gap_data, 1952), 1952)
```
